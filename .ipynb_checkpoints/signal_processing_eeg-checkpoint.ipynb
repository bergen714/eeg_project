```python
import mne
import matplotlib.pyplot as plt

# ─────────────────────────────────────────
# STEP 1 — Load the raw EDF file
# ─────────────────────────────────────────
raw = mne.io.read_raw_edf('files/S001/S001R01.edf', preload=True)

# Strip trailing dots from channel names so they match the 10-20 standard
# (PhysioNet pads names to 4 chars with dots, e.g. "Fc5." → "FC5", "C3.." → "C3")
# strip the channel names from their dots and make them upper case so it's standardized
raw.rename_channels({ch: ch.rstrip('.').upper() for ch in raw.ch_names})

# Apply standard 10-20 montage so electrode positions are known(?)
montage = mne.channels.make_standard_montage('standard_1005')
raw.set_montage(montage, on_missing='warn')

# ─────────────────────────────────────────
# STEP 2 — Explore basic info
# ─────────────────────────────────────────
print("=== BASIC INFO ===")
print(raw.info)

print("\n=== CHANNEL NAMES ===")
print(raw.ch_names)

print("\n=== KEY NUMBERS ===")
print(f"Sampling rate: {raw.info['sfreq']} Hz")
print(f"Number of channels: {len(raw.ch_names)}")
print(f"Duration: {raw.times[-1]:.1f} seconds")
print(f"Number of timepoints: {raw.n_times}")

# ─────────────────────────────────────────
# STEP 3 — Plot the raw signal
# ─────────────────────────────────────────
raw.plot(
    duration=10,
    n_channels=20,
    scalings='auto',
    title='Raw EEG — S001R01 (eyes open baseline)'
)
plt.show()

# ─────────────────────────────────────────
# STEP 4 — Plot sensor locations on head map
# ─────────────────────────────────────────
raw.plot_sensors(show_names=True)
plt.show()

# ─────────────────────────────────────────
# STEP 5 — Plot the power spectrum
# ─────────────────────────────────────────
raw.compute_psd(fmax=60).plot()
plt.show()

# ─────────────────────────────────────────
# STEP 6 — Load all 3 runs and compare
# ─────────────────────────────────────────
def load_and_fix(path):
    r = mne.io.read_raw_edf(path, preload=True)
    r.rename_channels({ch: ch.rstrip('.').upper() for ch in r.ch_names})
    r.set_montage(mne.channels.make_standard_montage('standard_1005'), on_missing='warn')
    return r

raw_eyes_open   = load_and_fix('files/S001/S001R01.edf')
raw_eyes_closed = load_and_fix('files/S001/S001R02.edf')
raw_motor       = load_and_fix('files/S001/S001R04.edf')

print("\n=== EYES OPEN ===")
print(f"Duration: {raw_eyes_open.times[-1]:.1f} seconds")
print(f"Annotations: {raw_eyes_open.annotations}")

print("\n=== EYES CLOSED ===")
print(f"Duration: {raw_eyes_closed.times[-1]:.1f} seconds")
print(f"Annotations: {raw_eyes_closed.annotations}")

print("\n=== MOTOR IMAGERY ===")
print(f"Duration: {raw_motor.times[-1]:.1f} seconds")
print(f"Annotations: {raw_motor.annotations}")

# ─────────────────────────────────────────
# STEP 7 — Compare power spectra across runs
# eyes closed should show more alpha (8-13 Hz)
# ─────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

raw_eyes_open.compute_psd(fmax=50).plot(axes=axes[0], show=False)
axes[0].set_title('Eyes Open')

raw_eyes_closed.compute_psd(fmax=50).plot(axes=axes[1], show=False)
axes[1].set_title('Eyes Closed')

raw_motor.compute_psd(fmax=50).plot(axes=axes[2], show=False)
axes[2].set_title('Motor Imagery')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib
#matplotlib.use('Agg')
import mne
import matplotlib.pyplot as plt
matplotlib.use('Agg')
from mne.datasets import eegbci
from mne.io import concatenate_raws, read_raw_edf

# Suppress MNE's verbose output — cleaner notebook
mne.set_log_level('WARNING')

print(f"MNE version: {mne.__version__}")
print("All imports OK.")
```

```python
raw_fnames = eegbci.load_data(
    subjects=1,
    runs=[6, 10, 14],
    path=''   # folder containing the PhysioNet data
)
raw = concatenate_raws([read_raw_edf(f, preload=True) for f in raw_fnames])

print("*** Raw EEG Information ***")

print(f'Channels: {len(raw.ch_names)} EEG channels.')
print(f'Sampling rate: {raw.info['sfreq']} Hz')
print(f'Duration: {raw.times[-1]:.1f} s ({raw.times[-1]/60:.1f} min)')
print(f"First 5 channels: {raw.ch_names[:5]}")
print()

# Set electrode positions (needed for ICA and topographic plots)
eegbci.standardize(raw)   # standardize channel names to 10-05 system
raw.set_montage('standard_1005')
print("Electrode montage set.")
```

## Visualize Raw Signal

```python
channels_to_plot = ['C3','C4','Cz','Fz']

# 5 seconds of sampling, if sampling rate is xHz, we have 5x sample points for our graph.
n_samples = int(5.0 * raw.info['sfreq'])
t = np.linspace(0,5.0,n_samples)

fig, axes = plt.subplots(4,1, figsize=(13,8),
                         sharex=True,facecolor='#F8F9FA')
fig.suptitle("Raw EEG — Before Preprocessing\nSubject 1, Motor Imagery Task",
             fontsize=12,fontweight='bold')
colors = ['#1E3A5F', '#9B1B3D', '#0F6E56', '#534AB7']

for ax, ch, color in zip(axes, channels_to_plot, colors):

    #find the index of the desired channel
    idx = raw.ch_names.index(ch)

    # extract a slice from the EEG data
    #from the channel at index idx
    #take the first row [0:,], n_samples -> the first few seconds we want to plot.
    #get the data of the channel with the idx index; then take its first row first five seconds
    data = raw.get_data(picks=[idx])[0,:n_samples] * 1e6  #V -> μV
    ax.plot(t, data, linewidth=0.8, color=color)
    ax.set_ylabel(f'{ch}\n(μV)', fontsize=9)
    ax.grid(True, alpha=0.3)

    #remove the thin black borders from top and right.
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)


    ax.text(0.01, 0.85, f'Range: {data.max() - data.min():.1f} μV', transform = ax.transAxes, fontsize=7, color='gray')
    axes[-1].set_xlabel('Time (s)', fontsize=10)
    plt.tight_layout()
    plt.savefig('eeg_raw_signal.png', dpi=150, bbox_inches='tight')
    plt.show()
    print("Saved eeg_raw_signal.png")
             
```

```python
from scipy.signal import welch

fig, ax = plt.subplots(figsize=(11,4), facecolor='#F8F9FA')

for ch, color in zip(['C3', 'C4'], ['#1E3A5F', '#9B1B3D']):

    idx  = raw.ch_names.index(ch)
    data = raw.get_data(picks=[idx])[0] * 1e6   # V → μV

    # The Welch method estimates the power spectral density (PSD), which is how much power
    # does the signal have at each frequency. (LEARN THE MATH BEHIND)
    # full-FFT: gives the overall image, therefore can cause us to see lots of unfamiliarities.
    # you dont know about the temporal dynamics, you cant see the more intricate picture.

    # Welch's method: isolate a little piece of the data. Average all the power spectra.
    # Divide the whole timeline into overlapping 512-sample windows, take the average of all 
    # and finally create the final spectrum. 
    # when a portion of the signal is on the edge of a segment, that edge gets mathematically distorted. Therefore,
    # having overlapping chunks will ensure that every piece of the overall recording is well-considered.

    # smallest frequency observed within a segment: 1/length_of_segment Hz
    # frequency resolution: how easy it is to distinguish between two different frequencies.
    # many segments lower the resolution, while few segments enable you to observe a wider range of frequencies.
    freq, psd = welch(data, fs=raw.info['sfreq'], nperseg=512)

    # plotted as a function of logarithm, because lower frequencies produce much more power than higher frequencies.
    ax.semilogy(freq, psd, color=color, linewidth=1.5, label=ch, alpha=0.9)


# Mark key frequencies
ax.axvline(60,  color='red',    linestyle='--', alpha=0.7, linewidth=1.5,
           label='60 Hz (powerline noise)')
ax.axvspan(8, 13, alpha=0.08, color='green', label='Alpha band (8–13 Hz)')

ax.set_xlabel('Frequency (Hz)', fontsize=10)
ax.set_ylabel('PSD (μV²/Hz)', fontsize=10)
ax.set_title('Raw Power Spectral Density — Before Filtering', fontsize=11)
ax.set_xlim(0, 80)
ax.legend(fontsize=9)
ax.grid(True, alpha=0.3)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('eeg_psd_raw.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: eeg_psd_raw.png")
print()
print("Note the 60 Hz spike — that's powerline interference.")
print("The notch filter will eliminate it.")
    
```

```python
# Make a copy so we preserve the original for comparison
raw_filtered = raw.copy()

# Bandpass filter: 0.5 Hz high-pass + 45 Hz low-pass
# fir_window='hamming' — same Hamming window you've used in signal processing
# keep the frequencies between 0.5Hz and 45Hz, throw everything else away.
raw_filtered.filter(
    l_freq=0.5,       # high-pass cutoff (Hz)
    h_freq=45.0,      # low-pass cutoff (Hz)
    fir_window='hamming', #smoother cut-off of limits.
    verbose=False
)

print("Bandpass filter applied: 0.5 – 45 Hz")
print(f"Filter type: FIR with Hamming window")
```

```python
# Notch filter at exactly 60 Hz
raw_filtered.notch_filter(
    freqs=60,
    verbose=False
)

print("Notch filter applied: 60 Hz removed")
print()
print("If this were a European dataset, you would use freqs=50.")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4), facecolor='#F8F9FA')
fig.suptitle('Filtering Effect — C3 Channel', fontsize=12, fontweight='bold')

for ax, data_source, title, color in zip(
    axes,
    [raw, raw_filtered],
    ['Raw (unfiltered)', 'After bandpass (0.5–45 Hz) + notch (60 Hz)'],
    ['#9B1B3D', '#1E3A5F']
):
    idx    = data_source.ch_names.index('C3')
    data   = data_source.get_data(picks=[idx])[0] * 1e6
    freqs, psd = welch(data, fs=data_source.info['sfreq'], nperseg=512)

    ax.semilogy(freqs, psd, color=color, linewidth=1.5)
    ax.axvline(60, color='red', linestyle='--', alpha=0.5, linewidth=1,
               label='60 Hz')
    ax.axvspan(8, 13, alpha=0.08, color='green', label='Alpha (8–13 Hz)')
    ax.set_xlim(0, 80)
    ax.set_xlabel('Frequency (Hz)', fontsize=9)
    ax.set_ylabel('PSD (μV²/Hz)', fontsize=9)
    ax.set_title(title, fontsize=10)
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('eeg_filter_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: eeg_filter_comparison.png")
print()
print("The 60 Hz spike should be gone in the right panel.")
print("The noise floor above 45 Hz should be flat/removed.")
```

## ICA Artifact Removal
- Decomposing the 64-channel EEG signal into independent components, which will all individually represent a different source of electrical activity in the body.
- Most are brain signals. But some of them also will be; eye blinks, eye movements, heart beat (ECG)
- With ICA, we will aim to remove the non-neural signals like above from the signal by their characteristic pattern and time course.
- In short, we will be seperating a mixed signal back to its original components, noise will be attempted to be removed.
- Every channel senses a combination of different sources of signals. ICA attempts to find the solutions for all signals with as many independent solutions as possible.

```python
from mne.preprocessing import ICA

raw_for_ica = raw_filtered.copy().filter(l_freq=1.0, h_freq=None, verbose=False)

# it will find 20 independent sources/components from the 64 channels. Think of it as solving for 20 variables in a 64 
# equation matrix. 64x20 matrix. These 20 independent artifcats could vary from core neural activity to eye blinks.
ica = ICA(n_components=20,
          random_state=42,
          max_iter='auto',
          verbose=False
         )

ica.fit(raw_for_ica)
print(f'ICA fitted. {ica.n_components_} found.')
```

```python
# Automatically detect eye blink components
# MNE uses the frontal channels (Fp1, Fp2) as reference since
# they're closest to the eyes and most contaminated by blinks

# eog_indicies: a list of component indicies which are flagged as likely eye blink activity from the 20 total components
# it's trying to dissemble to.
# eog_scores: a correlation list score per ICA component: one for Fp1 and one for Fp2.
eog_indices, eog_scores = ica.find_bads_eog(
    raw_filtered,
    ch_name=['Fp1', 'Fp2'],   # frontal channels as EOG proxy
    verbose=False
)

# which components got flagged.
print(f"Eye artifact components detected: {eog_indices}")
#2x20 matrix score list.
eog_scores_combined = np.abs(np.array(eog_scores)).max(axis=0)  # combine Fp1/Fp2 scores per component
print(f"Scores (correlation with eye channel): {[f'{s:.3f}' for s in eog_scores_combined[eog_indices]]}")

if len(eog_indices) == 0:
    print("No eye artifacts auto-detected.")
    print("This can happen with clean data — manually inspect components if needed.")
else:
    # Mark components for removal
    ica.exclude = eog_indices
    print(f"\nComponents {eog_indices} will be removed.")
```

```python
# Apply ICA — removes the marked components and reconstructs clean EEG
raw_clean = raw_filtered.copy()
ica.apply(raw_clean, verbose=False)

print("ICA artifact removal applied.")
print(f"Components removed: {ica.exclude}")
print(f"Components kept:    {ica.n_components_ - len(ica.exclude)} of {ica.n_components_}")
```

```python
# Compare raw vs clean at a channel typically contaminated by eye blinks
fig, axes = plt.subplots(2, 1, figsize=(13, 6), sharex=True, facecolor='#F8F9FA')
fig.suptitle('ICA Artifact Removal — Fp1 Channel (frontal, near eyes)',
             fontsize=12, fontweight='bold')

n_plot = int(10.0 * raw_clean.info['sfreq'])   # 10 seconds
t      = np.linspace(0, 10.0, n_plot)

for ax, source, title, color in zip(
    axes,
    [raw_filtered, raw_clean],
    ['Before ICA (eye blinks visible as large spikes)',
     'After ICA (eye blink components removed)'],
    ['#9B1B3D', '#1E3A5F']
):
    try:
        idx  = source.ch_names.index('Fp1')
    except ValueError:
        idx  = 0   # fallback if Fp1 not present
    data = source.get_data(picks=[idx])[0, :n_plot] * 1e6

    ax.plot(t, data, linewidth=0.8, color=color)
    ax.set_ylabel('Amplitude (μV)', fontsize=9)
    ax.set_title(title, fontsize=10)
    ax.grid(True, alpha=0.3)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

axes[-1].set_xlabel('Time (s)', fontsize=10)
plt.tight_layout()
plt.savefig('eeg_ica_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: eeg_ica_comparison.png")
print()
print("Large spikes in the top panel are eye blinks.")
print("These should be visibly reduced in the bottom panel.")
```

## Epoching
- Cuts the continuous EEG recording into short segments, each one time locked to a task event (stimulus or response)

In this dataset;

- T1 = left hand imagery
- T2 = right hand imagery.

Each one goes from -1.0s to 4.0s seconds around the stimulus event:

- `-1.0 to 0.0s` : baseline (prestimulus, at rest)
- `0.0 to 4.0s`: the task period.

```python
from mne import events_from_annotations, Epochs

# Extract event markers from the recording annotations
# events: each row is an event with three columns: sample_index, 0, event_id (what type of event: T0, T1, T2)
events, event_dict = events_from_annotations(raw_clean, verbose=False)

print("=== Events found ===")
print(f"Event types: {event_dict}")
print(f"Total events: {len(events)}")
print()

# Show event count per type
for event_name, event_id in event_dict.items():

    # take all rows and look at their event_ids. Add them to the sum if it equals to the desired event_id.
    count = np.sum(events[:, 2] == event_id)
    print(f"  {event_name}: {count} trials")
```

```python
# Creating the epochs

epochs = Epochs(
    raw_clean,
    events,
    event_id=event_dict,
    tmin=-1.0, #1 second before event
    tmax=4.0, #4 seconds after initial
    baseline=(-1.0,0.0), #setting the prestimulus period
    preload=True,
    verbose=False)


print("====Epochs created====")
print(f'Epoch window: {epochs.tmin} to {epochs.tmax} s')
print('Baseline period: -1.0 to 0.0 s.')
print(f"Total epochs:    {len(epochs)}")
print(f"Epoch duration:  {epochs.times[-1] - epochs.times[0]:.1f} s")
print(f"Samples/epoch:   {len(epochs.times)}")
print()
print(f"Epoch breakdown:")
for key in event_dict:
    if key in epochs.event_id:
        print(f"  {key}: {len(epochs[key])} epochs")
```

## Dropping bad epochs

reject any epochs > 300 μV

```python
# Count before rejection
n_before = len(epochs)

# Reject epochs with peak-to-peak amplitude > 150 μV in any channel
epochs.drop_bad(reject={'eeg': 300e-6}, verbose=False)   # 150 μV in Volts

n_after   = len(epochs)
n_dropped = n_before - n_after

print(f"=== Epoch rejection ===")
print(f"Before: {n_before} epochs")
print(f"After:  {n_after}  epochs")
print(f"Dropped: {n_dropped} epochs ({n_dropped/n_before*100:.1f}%)")
print()

# Split into condition-specific epoch sets
try:
    epochs_left  = epochs['T1']   # left hand imagery
    epochs_right = epochs['T2']   # right hand imagery
    print(f"Left hand (T1):  {len(epochs_left)} clean epochs")
    print(f"Right hand (T2): {len(epochs_right)} clean epochs")
except KeyError:
    # Some runs use different event labels
    keys = list(epochs.event_id.keys())
    epochs_left  = epochs[keys[0]]
    epochs_right = epochs[keys[1]]
    print(f"{keys[0]}: {len(epochs_left)} clean epochs")
    print(f"{keys[1]}: {len(epochs_right)} clean epochs")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 8), facecolor='#F8F9FA')
fig.suptitle('Epoched EEG — Motor Cortex Channels (C3, C4)',
             fontsize=12, fontweight='bold')

# the time axis for epochs (-1.0-4.0s)
t_epoch = epochs.times

# iterates over the rows. Each tuple within the list contains two things: epoch_data, label, and color
# enumerate wraps that list and gives you a counter (0 in the first and 1 in the second iteration, as there are two tuples)
for row, (ep_set, condition, cond_color) in enumerate([
    (epochs_left,  'Left hand imagery (T1)',  '#1E3A5F'),
    (epochs_right, 'Right hand imagery (T2)', '#9B1B3D'),
]):
    # iterates two times, one plot for C3 channel and the other for C4
    for col, ch in enumerate(['C3', 'C4']):

        #pick the right subplot (/epoch_left/ of /c3/ for example)
        ax = axes[row, col]
        try:

            #check if C3 or C4 exist in the channel list.
            ch_idx = ep_set.ch_names.index(ch)
        except ValueError:
            continue

        # pull all the epoch data for that one channel
        # a 2D-array where each row represents one trial.
        data = ep_set.get_data(picks=[ch_idx])[:, 0, :] * 1e6   # epochs × samples

        # Plot individual trials (faint)
        for trial in data[:min(10, len(data))]:   # show max 10 trials
            ax.plot(t_epoch, trial, color=cond_color, linewidth=0.4,
                    alpha=0.3)

        # Plot mean across trials (bold)
        ax.plot(t_epoch, data.mean(axis=0), color=cond_color,
                linewidth=2.0, label=f'Mean (n={len(data)})')

        ax.axvline(0, color='gray', linestyle='--', linewidth=1,
                   alpha=0.7, label='Event onset')
        ax.axvspan(-1, 0, alpha=0.05, color='gray', label='Baseline')
        ax.set_xlabel('Time (s)', fontsize=9)
        ax.set_ylabel('Amplitude (μV)', fontsize=9)
        ax.set_title(f'{condition} — {ch}', fontsize=10)
        ax.legend(fontsize=7)
        ax.grid(True, alpha=0.3)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.savefig('eeg_epochs.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: eeg_epochs.png")
```
